**Data Preprocessing**

```python
from sklearn.model_selection import train_test_split

X=df.drop(['Class'],axis=1)
y=df['Class']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20, random_state=44)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Applying SMOTE to the training set

from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=44)
X_train_sm, y_train_sm = sm.fit_resample(X_train, y_train)
```

```python
print(X_train_sm.value_counts().sum())
print(y_train_sm.value_counts().sum())
```

**Modeling**

```python
from sklearn.ensemble import RandomForestClassifier

# Applying RandomForestClassifier on the oversampled data
random_forest = RandomForestClassifier(n_estimators=30, n_jobs=-1, random_state=44)
random_forest.fit(X_train_sm, y_train_sm)

# predictions on the test set
y_pred = random_forest.predict(X_test)
```

**Report**

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
cm = np.array(confusion_matrix(y_test, y_pred, labels=[1,0]))
confusion = pd.DataFrame(cm, index=['is Fraud', 'is legit'],columns=['predicted fraud','predicted legit'])
confusion
```

```python
from sklearn.metrics import accuracy_score, recall_score, confusion_matrix,roc_auc_score,precision_score,f1_score,classification_report
import seaborn as sns
from matplotlib import pyplot

lr_accuracy = accuracy_score(y_test, y_pred)
lr_recall = recall_score(y_test, y_pred)
lr_cm = confusion_matrix(y_test, y_pred)
lr_auc = roc_auc_score(y_test, y_pred)
lr_precision = precision_score(y_test, y_pred)
lr_f1 = f1_score(y_test, y_pred)
lr_cl = classification_report(y_test, y_pred)

print("Accuracy of Logistic Regression: {:.4%}".format(lr_accuracy))
print("Recall: ",(lr_recall))
print("ROC AUC: ",(lr_auc))
print("Precision score: ",(lr_precision))
print("F1 score: ",(lr_f1))
print("Confusion matrix:\n",(lr_cm))
print("Classification report:\n",(lr_cl))

lr_cm = pd.DataFrame(lr_cm, ['True Normal','True Fraud'],['Prediction Normal','Prediction Fraud'])
pyplot.figure(figsize = (8,4))
sns.set(font_scale=1.4)
sns.heatmap(lr_cm, annot=True,annot_kws={"size": 16},fmt='g')
```
